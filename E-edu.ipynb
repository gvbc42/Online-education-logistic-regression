                             not-converted     converted

              converted           FN             TP

```python
metrics.accuracy_score(y_train_pred_final.Converted,y_train_pred_final.Predicted)
```

```python
# Define terms for sensitivity
TN=confusion_metrx[0,0]
FP=confusion_metrx[0,1]
FN=confusion_metrx[1,0]
TP=confusion_metrx[1,1]
```

```python
# Sensitivity (P) calculation
P=TP/(TP+FN)
P
```

```python
# Specifitivity(R)
R=TN/(TN+FP)
R
```

```python
# Define ROC curve for 
def plot_roc(act,prob):
    fp,tp, threshold=metrics.roc_curve(act,prob,drop_intermediate=False)
    auc_score=metrics.roc_auc_score(act,prob)
    plt.figure(figsize=(4,4))
    plt.plot(fp,tp,label="ROC Plot (area=%0.2f)" % auc_score)
    plt.plot([0,1],[0,1],"k--")
    plt.xlim(0.0,1.0)
    plt.ylim(0.0,1.06)
    plt.legend(loc="upper left")
    plt.show()

    return None
```

```python
fp,tp, threshold=metrics.roc_curve(y_train_pred_final.Converted,y_train_pred_final.Conversion_prob)
```

```python
plot_roc(y_train_pred_final.Converted,y_train_pred_final.Conversion_prob)
```

```python
cut_df=pd.DataFrame(columns=["Probability","Accuracy","Sencitivity_p","Specivicity_R"])
```

```python
numbers
```

```python
for i in numbers:
    cm1=metrics.confusion_matrix(y_train_pred_final["Converted"],y_train_pred_final[i])
    total1=sum(sum(cm1))
    accuracy=(cm1[0,0]+cm1[1,1])/total1
    specificity=(cm1[0,0])/(cm1[0,0]+cm1[0,1])
    sencitvity=(cm1[1,1])/(cm1[1,1]+cm1[1,0])
    cut_df.loc[i]=[i,accuracy,sencitvity,specificity]
```
